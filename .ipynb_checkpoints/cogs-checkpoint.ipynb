`packages`

```python
import pandas as pd
import numpy  as np
import glob, sys, psycopg2
from sqlalchemy import create_engine
#import pandas_montecarlo  -- package for Monte Carlo simulations
```

`functions`

```python
def read_cogs(paths):
    cogs_list_ = []    
    
    for file_ in paths:
        df = pd.read_excel(file_, sheet_name = "RackPrice")              
        cogs_list_.append(df)
        print(file_)
        
    cogs_frame = pd.concat(cogs_list_, axis = 0, ignore_index = True, 
                           sort = False)
    return cogs_frame
```

`main`

```python
source_path = r"C:\Users\migue\OneDrive\Documentos\BP\datasets\Costos\RackPrice\Rack*.xlsx"

structure = {"Region":[], "Regular":[], "Premium":[], "Diesel":[], "Fecha":[]}
cogs_df = pd.DataFrame(data = structure)

path_cogs = glob.glob(source_path)    
cogs = read_cogs(path_cogs)

cogs = cogs.rename(columns = {"REGIN" : "Terminal", 
                              "Precio TAR-regular": "Precio TAR Regular (m3)", 
                              "Precio TAR-premium": "Precio TAR Premium (m3)", 
                              "Precio TAR-diesel": "Precio TAR Diesel (m3)",
                              "fecha" : "Fecha"})

cogs["Precio TAR Regular (lt)"] = cogs["Precio TAR Regular (m3)"].apply(lambda x : x/1000)
cogs["Precio TAR Premium (lt)"] = cogs["Precio TAR Premium (m3)"].apply(lambda x : x/1000)
cogs["Precio TAR Diesel (lt)"] = cogs["Precio TAR Diesel (m3)"].apply(lambda x : x/1000)

cogs["Fecha"] = cogs.Fecha.astype('datetime64')
cogs["Terminal"] = cogs["Terminal"].str.replace('', 'A', regex = False).str.replace('', 'E').str.replace('', 'I').str.replace('', 'O').str.replace('', 'U').str.replace('', 'N')
```

```python
cogs.drop(columns=['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8'], inplace=True)
```

```python
# postgresql key
usr = "postgres"
pwd = "V3lazco_1989"
db =  "kaggle_data"
port = "5432"

# create the engine for the postgresql object
engine = create_engine("postgresql://" + usr + ":" + pwd + "@localhost:5432/" + db)
```
